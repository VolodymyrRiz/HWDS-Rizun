```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Створення трьох векторів розміром 4 з довільними цілими числами від 1 до 50
vector1 = np.random.randint(1, 51, size=4)
vector2 = np.random.randint(1, 51, size=4)
vector3 = np.random.randint(1, 51, size=4)

# Створення 3D графіка
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Виведення графічно векторів
ax.scatter(vector1, vector2, vector3, c='r', marker='o')
ax.set_xlabel('Vector 1')
ax.set_ylabel('Vector 2')
ax.set_zlabel('Vector 3')
ax.set_title('3D Scatter Plot of Vectors')

plt.show()
```
