```python
%matplotlib inline
import pandas as pd # type: ignore
import numpy as np # type: ignore
import matplotlib.pyplot as plt
import seaborn as sns

# Завантажте CSV-файл у DataFrame pandas з використанням кодування cp1251
table = pd.read_csv('2017_jun_final.csv')

# Виведіть перші рядки таблиці
print('_________________________________________________________')
print("ПЕРШІ РЯДКИ ТАБЛИЦІ")
print(table.head(5))        

# Виведіть інформацію про таблицю
print('_________________________________________________________')
print("ІНФОРМАЦІЯ ПРО ТАБЛИЦЮ:")
rows, columns = table.shape
print(f"Кількість рядків: {rows}")
print(f"Кількість стовпців: {columns}")

# Виведіть типи даних кожного стовпця
print('Типи всіх стовпців:')
print(table.dtypes)

# Обчисліть відсоток пропусків в кожному стовпці після заміни
missing_percentages = table.isnull().sum() / len(table) * 100

# Виведіть інформацію про таблицю
print('_________________________________________________________')
print("ІНФОРМАЦІЯ ПРО ТАБЛИЦЮ.")
print('Частка пропусків, що міститься в кожній колонці: ')
print(missing_percentages)

# Замініть всі пропуски на NaN
table.replace('', np.nan, inplace=True)

# Виведіть оновлену таблицю
print('_________________________________________________________')
print("ТАБЛИЦЯ ПІСЛЯ ЗАМІНИ ПРОПУСКІВ НА NaN:")
print(table.head(5))


# Визначте назви колонок, які треба вилучити
columns_to_drop = [3, 14, 17, 23, 24]  # Віднімаємо одиницю, оскільки індексація починається з 0

# Вилучіть вказані колонки
table_updated = table.drop(table.columns[columns_to_drop], axis=1)

# Виведіть оновлену таблицю
print('_________________________________________________________')
print("ТАБЛИЦЯ ПІСЛЯ ВИДАЛЕННЯ ВИБРАНИХ СТОВПЦІВ:")
print(table_updated.head(5))

# Обчисліть відсоток пропусків в кожному стовпці після заміни
missing_percentages = table_updated.isnull().sum() / len(table) * 100

# Виведіть інформацію про таблицю
print('_________________________________________________________')
print("ІНФОРМАЦІЯ ПРО ТАБЛИЦЮ.")
print('Частка пропусків, що міститься в кожній колонці після видалення стовпців, окрім МОВА ПРОГРАМУВАННЯ: ')
print(missing_percentages)

import seaborn as sns
import matplotlib.pyplot as plt

# Діаграма для відображення відсотків пропусків
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_percentages.index, y=missing_percentages.values, palette="Greens")
plt.title('Відсоток пропусків у кожній колонці')
plt.xlabel('Колонка')
plt.ylabel('Відсоток пропусків')
plt.xticks(rotation=90)  # Для зручності можна повернути мітки на 90 градусів
plt.show()

# Створіть нову таблицю python_data, яка містить тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = table_updated[table_updated[table_updated.columns[2]] == 'Python']

# Виведіть нову таблицю python_data
print('_________________________________________________________')
print("ТАБЛИЦЯ З РЯДКАМИ, ЯКІ ВИБРАЛИ МОВУ ПРОГРАМУВАННЯ 'Python':")
print(python_data)

# Виведіть інформацію про таблицю python_data
print('_________________________________________________________')
print("ІНФОРМАЦІЯ ПРО ТАБЛИЦЮ python_data:")
rows, columns = python_data.shape
print(f"Кількість рядків: {rows}")
print(f"Кількість стовпців: {columns}")

# Видаліть всі рядки з таблиці table_updated
table_updated = table_updated.drop(table_updated.index)

# Виведіть оновлену таблицю
print('_________________________________________________________')
print("ВИХІДНА ТАБЛИЦЯ - не  python_data - ПІСЛЯ ВИДАЛЕННЯ ВСІХ РЯДКІВ:")
print(table_updated)

# Виведіть інформацію про таблицю після видалення всіх рядків
print('_________________________________________________________')
print("ІНФОРМАЦІЯ ПРО ТАБЛИЦЮ:")
rows, columns = table_updated.shape
print(f"Кількість рядків: {rows}")
print(f"Кількість стовпців: {columns}")

# Виконайте групування за стовпчиком "Посада" для таблиці python_data
grouped_data = python_data.groupby('Посада')

# Виведіть групувані дані
print('_________________________________________________________')
print("ГРУПУВАННЯ ЗА СТОВПЦЕМ 'ПОСАДА' ДЛЯ ТАБЛИЦІ python_data:")
print(grouped_data.first())  # Виводимо перший елемент кожного групи

# Виведіть інформацію про групувані дані
print('_________________________________________________________')
print("ІНФОРМАЦІЯ ПРО ГРУПИ:")
for name, group in grouped_data:
    print(f"Група '{name}' має {len(group)} елементів")
    
# Знайдіть мінімальне та максимальне значення у стовпчику "Зарплата на місяць" для кожної групи
aggregated_data = grouped_data['Зарплата на місяць'].agg(['min', 'max'])

# Виведіть агреговані дані
print('_________________________________________________________')
print("АГРЕГАЦІЯ ЗА СТОВПЦЕМ 'ПОСАДА' ДЛЯ ТАБЛИЦІ python_data:")
print(aggregated_data)


# Діаграма для відображення максимальної та мінімальної заробітної плати за посадою
plt.figure(figsize=(10, 6))
sns.lineplot(data=aggregated_data, palette="Reds")
plt.title('Максимальна та мінімальна заробітна плата за посадою')
plt.xlabel('Посада')
plt.ylabel('Заробітна плата')
plt.xticks(rotation=90)  # Для зручності можна повернути мітки на 90 градусів
plt.show()

# Функція для обчислення середньої заробітної плати
def fill_avg_salary(row):
    return row.mean()

# Застосування функції fill_avg_salary до кожного рядка DataFrame aggregated_data
aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)

# Виведіть агреговані дані з новим стовпчиком "avg"
print('_________________________________________________________')
print("АГРЕГАЦІЯ ЗА СТОВПЦЕМ 'ПОСАДА' ДЛЯ ТАБЛИЦІ python_data З НОВИМ СТОВПЦЕМ 'avg':")
print(aggregated_data)

# Діаграма для відображення середньої заробітної плати
plt.figure(figsize=(10, 6))
sns.barplot(x=aggregated_data.index, y=aggregated_data['avg'], palette="Blues")
plt.title('Середня заробітна плата за посадою')
plt.xlabel('Посада')
plt.ylabel('Середня заробітна плата')
plt.xticks(rotation=90)  # Для зручності можна повернути мітки на 90 градусів
plt.show()

# Збережіть DataFrame aggregated_data у файл CSV
aggregated_data.to_csv('aggregated_data1.csv', index=False)

# Створення описової статистики для стовпчика "avg"
avg_stats = aggregated_data['avg'].describe()

# Виведіть описову статистику для стовпчика "avg"
print('_________________________________________________________')
print("ОПИСОВА СТАТИСТИКА ДЛЯ СТОВПЦЯ 'avg':")
print(avg_stats)

# Збережіть DataFrame python_data у файл CSV
python_data.to_csv('python_data.csv', index=False)





```
