```python
%matplotlib inline
import pandas as pd # type: ignore
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
import warnings
import seaborn as sns # type: ignore

warnings.filterwarnings("ignore", category=FutureWarning)

# Завантажте CSV-файл у DataFrame pandas з використанням кодування cp1251
table = pd.read_csv('bestsellers with categories.csv')
table.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

# Виведіть перші рядки таблиці
print('_________________________________________________________')
print("ПЕРШІ РЯДКИ ТАБЛИЦІ")
print(table.head(5))        

# Виведіть інформацію про таблицю
print('_________________________________________________________')
print("ІНФОРМАЦІЯ ПРО ТАБЛИЦЮ:")
rows, columns = table.shape
print(f"Кількість рядків: {rows}")
print(f"Кількість стовпців: {columns}")

# Унікальні книги
unique_books = table.drop_duplicates(subset=['name', 'author'])

# Виведіть кількість унікальних книг
print('_________________________________________________________')
print("КІЛЬКІСТЬ УНІКАЛЬНИХ КНИГ:")
print(f"Кількість унікальних книг: {len(unique_books)}")

# Кількість пропусків в кожному стовпці
missing = table.isna().sum()

# Збільшення максимальної кількості рядків, які будуть виводитися
pd.set_option('display.max_rows', None)

# Виведіть інформацію про таблицю
print('_________________________________________________________')
print("ІНФОРМАЦІЯ ПРО ТАБЛИЦЮ.")
print('Кількість пропусків, що міститься в кожній колонці: ')
print(missing)

# Перевірте унікальні значення в колонці genre
unique_genres = table['genre'].unique()

# Виведіть унікальні значення
print('_________________________________________________________')
print("УНІКАЛЬНІ ЗНАЧЕННЯ В КОЛОНЦІ ЖАНРУ:")
print(unique_genres)

# Побудова діаграми розподілу цін
plt.figure(figsize=(10, 6))
table['price'].hist(bins=30, edgecolor='black')
plt.title('Розподіл цін')
plt.xlabel('Ціна')
plt.ylabel('Кількість')
plt.show()

# Максимальна ціна
max_price = table['price'].max()

# Мінімальна ціна
min_price = table['price'].min()

# Середня ціна
mean_price = table['price'].mean()

# Медіанна ціна
median_price = table['price'].median()

# Виведіть статистичні характеристики
print('_________________________________________________________')
print("СТАТИСТИЧНІ ЧАСТИНИ ЦІН:")
print(f"Максимальна ціна: {max_price}")
print(f"Мінімальна ціна: {min_price}")
print(f"Середня ціна: {mean_price}")
print(f"Медіанна ціна: {median_price}")

# Діаграма розмаху для статистичних характеристик цін
plt.figure(figsize=(10, 6))
sns.boxplot(x=table['price'], y=table['category'], palette="Greens")
plt.title('Діаграма розмаху для статистичних характеристик цін')
plt.xlabel('Ціна')
plt.show()

# Найвищий рейтинг
max_rating = table['user_rating'].max()

# Виведіть найвищий рейтинг
print('_________________________________________________________')
print("НАЙВИЩИЙ РЕЙТИНГ У ДАТАСЕТІ:")
print(f"Найвищий рейтинг: {max_rating}")

# Знайдіть найвищий рейтинг
max_rating = table['user_rating'].max()

# Знайдіть скільки книг мають найвищий рейтинг
books_with_max_rating = table['user_rating'].value_counts()[max_rating]

# Виведіть кількість книг з найвищим рейтингом
print('_________________________________________________________')
print("КІЛЬКІСТЬ КНИГ З НАЙВИЩИМ РЕЙТИНГОМ:")
print(f"Кількість книг з найвищим рейтингом ({max_rating}): {books_with_max_rating}")

# Знайдіть книгу з найбільшою кількістю відгуків
book_with_most_reviews = table.loc[table['reviews'].idxmax(), 'name']

# Виведіть назву книги з найбільшою кількістю відгуків
print('_________________________________________________________')
print("КНИГА З НАЙБІЛЬШОЮ КІЛЬКІСТЬЮ ВІДГУКІВ:")
print(f"Книга з найбільшою кількістю відгуків: {book_with_most_reviews}")

# Створіть новий DataFrame top50_2015, який містить тільки книги з 2015 року
top50_2015 = table[table['year'] == 2015]

# Знайдіть книгу з найбільшою ціною у Топ-50 2015 року
most_expensive_book_2015 = top50_2015.loc[top50_2015['price'].idxmax(), 'name']

# Виведіть назву книги з найбільшою ціною у Топ-50 2015 року
print('_________________________________________________________')
print("КНИГА З НАЙБІЛЬШОЮ ЦІНОЮ У ТОП-50 2015 РОКУ:")
print(f"Книга з найбільшою ціною у Топ-50 2015 року: {most_expensive_book_2015}")

# Визначте кількість книг жанру Fiction, які потрапили до Топ-50 у 2010 році
fiction_books_2010 = table[(table['genre'] == 'Fiction') & (table['year'] == 2010)]

# Виведіть кількість книг жанру Fiction, які потрапили до Топ-50 у 2010 році
print('_________________________________________________________')
print("КІЛЬКІСТЬ КНИГ ЖАНРУ FICTION, ЯКІ ПОТРАПИЛИ ДО ТОП-50 У 2010 РОЦІ:")
print(f"Кількість книг жанру Fiction, які потрапили до Топ-50 у 2010 році: {len(fiction_books_2010)}")

# Визначте кількість книг з рейтингом 4.9, які потрапили до Топ-50 у 2010 та 2011 роках
books_with_rating_49 = table[(table['user_rating'] == 4.9) & (table['year'].isin([2010, 2011]))]

# Виведіть кількість книг з рейтингом 4.9, які потрапили до Топ-50 у 2010 та 2011 роках
print('_________________________________________________________')
print("КІЛЬКІСТЬ КНИГ З РЕЙТИНГОМ 4.9, ЯКІ ПОТРАПИЛИ ДО ТОП-50 У 2010 ТА 2011 РОКАХ:")
print(f"Кількість книг з рейтингом 4.9, які потрапили до Топ-50 у 2010 та 2011 роках: {len(books_with_rating_49)}")

# Відфільтруйте книги, які потрапили до рейтингу в 2015 році і коштують дешевше за 8 доларів
filtered_books = table[(table['year'] == 2015) & (table['price'] < 8)]

# Відсортуйте книги за зростанням ціни
sorted_books = filtered_books.sort_values(by='price', ascending=True)

# Виведіть відсортовані книги
print('_________________________________________________________')
print("ВІДСОРТОВАНІ КНИГИ З НАЙНИЖЧОЮ ЦІНОЮ:")
print(sorted_books[['name', 'price']])

# Групування за жанром та знаходження максимальної та мінімальної ціни для кожного жанру
price_ranges = table.groupby('genre').agg({'price': ['min', 'max']})

# Виведення результатів
print('_________________________________________________________')
print("МАКСИМАЛЬНІ ТА МІНІМАЛЬНІ ЦІНИ ДЛЯ КОЖНОГО ЖАНРУ:")
print(price_ranges)

# Діаграма лінійчата для максимальних та мінімальних цін для кожного жанру
# Перед тим, як побудувати лінійчку, вам потрібно перетворити price_ranges на правильний формат
price_ranges_long = price_ranges.stack().reset_index()
price_ranges_long.columns = ['Жанр', 'Тип', 'Ціна']

plt.figure(figsize=(10, 6))
sns.lineplot(data=price_ranges_long, x='Жанр', y='Ціна', hue='Тип', style='Тип')
plt.title('Лінійчата діаграма для максимальних та мінімальних цін для кожного жанру')
plt.xlabel('Жанр')
plt.ylabel('Ціна')
plt.legend(title='Тип ціни')
plt.xticks(rotation=45)  # Для зручності можна повернути мітки на 45 градусів
plt.show()

# Групування за автором та підрахунок кількості книг для кожного автора
book_counts = table.groupby('author').agg({'name': 'count'})

# Виведення результатів
print('_________________________________________________________')
print("КІЛЬКІСТЬ КНИГ ДЛЯ КОЖНОГО АВТОРА:")
print(book_counts)

# Групування за автором та підрахунок середнього рейтингу для кожного автора
average_ratings = table.groupby('author').agg({'user_rating': 'mean'})

# Виведення результатів
print('_________________________________________________________')
print("СЕРЕДНІ РЕЙТИНГИ ДЛЯ КОЖНОГО АВТОРА:")
print(average_ratings)

# З'єднання двох DataFrame за їх індексами (авторами)
author_stats = pd.concat([book_counts, average_ratings], axis=1)

# Перейменування стовпців
author_stats.columns = ['Кількість книг', 'Середній рейтинг']

# Виведення результатів
print('_________________________________________________________')
print("СТАТИСТИКА ДЛЯ КОЖНОГО АВТОРА:")
print(author_stats)

# Відсортування DataFrame за зростаючою кількістю книг та зростаючим рейтингом
author_stats_sorted = author_stats.sort_values(by=['Кількість книг', 'Середній рейтинг'], ascending=[True, True])

# Виведення результатів
print('_________________________________________________________')
print("СТАТИСТИКА за зростаючою кількістю книг та зростаючим рейтингом ДЛЯ КОЖНОГО АВТОРА:")
print(author_stats_sorted)


# Відповідь: Про скільки книг зберігає дані датасет? Про 351 книгу
# Відповідь: Чи є в якихось змінних пропуски? (Так / ні) НІ
# Відповідь: Які є унікальні жанри? ['Non Fiction' 'Fiction']
# Відповідь: Максимальна ціна? 105
# Відповідь: Мінімальна ціна? 0
# Відповідь: Середня ціна? 13.1
# Відповідь: Медіанна ціна? 11.0

# Відповідь: Який рейтинг у датасеті найвищий? Відповідь: 4.9
# Відповідь: Скільки книг мають такий рейтинг? Відповідь: 52
# Відповідь: Яка книга має найбільше відгуків? Відповідь: Where the Crawdads Sing
# Відповідь: З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча (можна використати проміжний датафрейм)? Відповідь: Diagnostic and Statistical Manual of Mental Disorders, 5th Edition: DSM-5
# Відповідь: Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році 
# (використовуйте &)? Відповідь: 20
# Відповідь: Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках 
# (використовуйте | або функцію isin)? Відповідь: 1
# Відповідь: Яка книга остання у відсортованому списку? Відповідь: Old School (Diary of a Wimpy Kid #10)

# Для початку давайте подивимося на максимальну та мінімальну ціни для кожного з жанрів (використовуйте функції groupby та agg, для підрахунку мінімальних та максимальних значень використовуйте max та min). Не беріть усі стовпці, виберіть тільки потрібні вам
# Відповідь: Максимальна ціна для жанру Fiction: Відповідь: 82
# Відповідь: Мінімальна ціна для жанру Fiction: Відповідь: 0
# Відповідь: Максимальна ціна для жанру Non Fiction: Відповідь: 105
# Відповідь: Мінімальна ціна для жанру Non Fiction: Відповідь: 0
# Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів 
# (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count). 
# Не беріть усі стовпці, виберете тільки потрібні
# Відповідь: Якої розмірності вийшла таблиця? Відповідь: 248 рядків та 1 стовпець
# Відповідь: Який автор має найбільше книг? Відповідь: Jeff Kinney
# Відповідь: Скільки книг цього автора? Відповідь: 12
# Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора 
# (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте 
# mean). Не беріть усі стовпці, виберете тільки потрібні
# Відповідь: У якого автора середній рейтинг мінімальний? Відповідь: Donna Tartt
# Відповідь: Який у цього автора середній рейтинг? Відповідь: 3.9
# З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг 
# та середній рейтинг (Використовуйте функцію concat з параметром axis=1). 
# Збережіть результат у змінну
# Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте 
# функцію sort_values)
# Відповідь: Який автор перший у списку? Відповідь: Muriel Barbery
```
